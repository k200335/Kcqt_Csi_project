```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium.common.exceptions import UnexpectedAlertPresentException
import re
import time

# 1. 브라우저 옵션 설정
options = Options()
options.add_experimental_option("detach", True)

# 2. 드라이버 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

try:
    # 3. 사이트 접속
    driver.get("https://www.csi.go.kr/index.do")
    
    print("\n" + "="*70)
    print(" [작업 지시]")
    print(" 1. 크롬 창에서 로그인을 완료하세요.")
    print(" 2. '품질시험 의뢰서 상세' 페이지(전체 내용이 보이는 곳)로 이동하세요.")
    print(" 3. 모든 표가 화면에 다 떴다면, 여기서 Enter를 누르세요.")
    print("="*70 + "\n")
    
    input("준비가 완료되었다면 Enter를 누르세요...")

    # 4. 전체 데이터 영역(#rqstViewAjax) 지정
    # 이 안에 의뢰내용, 시료정보, 공사개요 테이블이 모두 들어있습니다.
    container = driver.find_element(By.ID, "rqstViewAjax")
    all_tables = container.find_elements(By.TAG_NAME, "table")

    print(f"\n[검색 결과: 총 {len(all_tables)}개의 데이터 테이블을 발견했습니다.]\n")

    for i, table in enumerate(all_tables, 1):
        # 테이블의 용도를 알기 위해 바로 위의 제목(h5 등)을 찾아보려 시도합니다.
        try:
            title = table.find_element(By.XPATH, "./preceding-sibling::h5[1] | ../preceding-sibling::h5[1]").text
        except:
            title = f"미분류 테이블 {i}"

        print(f"\n▶ {title}")
        print("-" * 50)

        rows = table.find_elements(By.TAG_NAME, "tr")
        for row in rows:
            # th(제목칸)와 td(내용칸)를 모두 가져와서 합칩니다.
            cells = row.find_elements(By.XPATH, "./th | ./td")
            
            # 각 칸의 텍스트를 추출 (줄바꿈은 공백으로 처리)
            row_data = [cell.text.strip().replace('\n', ' ') for cell in cells if cell.text.strip()]
            
            if row_data:
                # 데이터가 존재하면 ' : ' 구분자로 보기 좋게 출력
                print(" | ".join(row_data))
        
    print("\n" + "="*70)
    print("전체 페이지 데이터 추출 테스트가 완료되었습니다.")

except Exception as e:
    print(f"\n[오류 발생] 데이터를 가져오는 중 에러가 났습니다: {e}")

# 브라우저를 닫지 않고 유지 (결과 확인용)


```
