```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium.common.exceptions import UnexpectedAlertPresentException
import re
import time

options = Options()
options.add_experimental_option("detach", True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

try:
    driver.get("https://gcloud.csi.go.kr/cmq/main.do")
    print("\n" + "="*70)
    print(" 1. 로그인 후 '상세 페이지'로 이동하여 빨간 테두리 내용이 화면에 보이게 하세요.")
    print(" 2. 화면 로딩이 완전히 끝나면 여기서 Enter를 누르세요.")
    print("="*70 + "\n")
    input("준비 완료 후 Enter...")

    # [수정 포인트] 특정 테이블을 지정하지 않고, 전체 컨테이너 안의 모든 행(tr)을 직접 타겟팅합니다.
    container = driver.find_element(By.ID, "rqstViewAjax")
    all_rows = container.find_elements(By.TAG_NAME, "tr")

    print(f"\n[데이터 스캔 결과: 총 {len(all_rows)}개의 행(Row) 탐색]")
    print("-" * 70)

    for idx, row in enumerate(all_rows):
        # 해당 줄(tr) 내부에 있는 모든 하위 요소의 텍스트를 가져옵니다.
        # 이렇게 하면 th, td뿐만 아니라 내부 span 등에 숨은 텍스트도 가져옵니다.
        row_text = row.text.strip().replace('\n', ' | ')
        
        if row_text:
            print(f"Line {idx+1}: {row_text}")

    print("\n" + "-" * 70)
    print("빨간 테두리 영역을 포함한 전체 스캔이 완료되었습니다.")

except Exception as e:
    print(f"오류 발생: {e}")

```
