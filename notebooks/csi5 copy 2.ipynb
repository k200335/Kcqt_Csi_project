```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium.common.exceptions import UnexpectedAlertPresentException
import re
import time

options = Options()
options.add_experimental_option("detach", True)
# options.add_argument("--headless")
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)




try:
    # 3. 사이트 접속
    driver.get("https://www.csi.go.kr/index.do")
    
    print("\n" + "="*60)
    print(" [주의] 아래 순서를 따라주세요:")
    print(" 1. 열린 크롬 창에서 직접 로그인을 진행합니다.")
    print(" 2. '품질시험 의뢰서 상세' 페이지(표가 있는 화면)까지 이동합니다.")
    print(" 3. 화면이 다 준비되었다면, 여기(터미널)에서 Enter를 누르세요.")
    print("="*60 + "\n")
    
    input("페이지 준비 완료 후 Enter를 누르세요...")

    # 4. 데이터 추출 (스크린샷에 나타난 CSS 선택자 적용)
    # id="rqstViewAjax" 안에 있는 모든 테이블 줄(tr)을 가져옵니다.
    rows = driver.find_elements(By.CSS_SELECTOR, "#rqstViewAjax tr")
    
    print("\n[추출된 표 내용]")
    print("-" * 40)
    for row in rows:
        # 각 줄의 텍스트가 비어있지 않은 경우에만 출력
        text = row.text.strip()
        if text:
            print(text)
    print("-" * 40)

except Exception as e:
    print(f"\n데이터 추출 중 오류 발생: {e}")



```
