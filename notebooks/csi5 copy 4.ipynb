```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium.common.exceptions import UnexpectedAlertPresentException
import re
import time


options = Options()
options.add_experimental_option("detach", True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

try:
    driver.get("https://www.csi.go.kr/index.do")
    print("\n" + "="*70)
    print(" 1. 로그인 후 '상세 페이지' 전체가 보이도록 하세요.")
    print(" 2. 준비가 되었다면 여기서 Enter를 누르세요.")
    print("="*70 + "\n")
    input("페이지 로딩 완료 후 Enter...")

    # 방법 1: 가장 넓은 범위인 'contentsRow' 클래스 내부를 전체 탐색
    # (이미지상에서 이 클래스가 모든 내용을 포함하는 부모로 보입니다)
    print("\n[전체 데이터 스캔 시작...]")
    print("-" * 70)

    # 1단계: 모든 테이블 데이터 추출
    main_content = driver.find_element(By.CLASS_NAME, "contentsRow")
    tables = main_content.find_elements(By.TAG_NAME, "table")
    
    for i, table in enumerate(tables, 1):
        print(f"\n[테이블 #{i}]")
        rows = table.find_elements(By.TAG_NAME, "tr")
        for row in rows:
            # 텍스트가 있는 모든 셀(th, td) 추출
            cells = row.find_elements(By.XPATH, ".//*[self::th or self::td]")
            row_text = [c.text.strip().replace('\n', ' ') for c in cells if c.text.strip()]
            if row_text:
                print(" | ".join(row_text))

    # 2단계: 테이블 외부에 개별적으로 존재하는 텍스트 (공사개요 등) 보완 추출
    print("\n\n[기타 항목 및 라벨 스캔]")
    others = main_content.find_elements(By.XPATH, ".//div[contains(@class, 'title')] | .//h5")
    for item in others:
        if item.text.strip():
            print(f"> {item.text.strip()}")

    print("\n" + "-" * 70)
    print("스캔이 완료되었습니다.")

except Exception as e:
    print(f"오류 발생: {e}")
```
